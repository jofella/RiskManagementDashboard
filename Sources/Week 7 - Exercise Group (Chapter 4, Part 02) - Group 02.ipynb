# Topic 4 - Estimating the Parameters of a $t$-Copula

In the file `t_copula.csv` you find a collection of 5000 bivariate random vectors that are distributed according to a $t$-copula $C^t_{\nu, R}$ with uniform marginals. Our goal is to estimate the parameters $\nu > 0$ and $R \in \mathbb{R}^{2 \times 2}$ of this copula using the methods described in Section 4.3.5. To perform the estimation, you find a function `kendall_tau(X)` in the code snippet below that computes Kendall's Tau $\widehat \varrho_\tau$ from an array of bivariate data of shape `(n, 2)`.

```python
def kendall_tau(X):
    n = X.shape[0]
    tau = np.zeros(n - 1)
    
    for k in range(n - 1):
        tau[k] = np.sign((X[k, 0] - X[k+1:, 0]) * (X[k, 1] - X[k+1:, 1])).sum()
    
    return tau.sum() / binom(n, 2)
```
